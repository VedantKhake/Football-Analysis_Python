```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns

```

```python
df=pd.read_csv(r"C:\Users\vedan\OneDrive\Desktop\Python viz\erling_haaland_2022_understat.csv")
```

```python
df
```

```python
df['x']=df['x']*120
df['y']=df['y']*80
```

```python
df
```

```python

# Filter only Haaland's shots against West Ham
df_filtered = df[df['h_team'] == 'West Ham']

fig, ax = plt.subplots(figsize=(13.5, 6))
fig.set_facecolor('#22312b')  # Set figure background
ax.patch.set_facecolor('#22312b')  # Set axis background

# Draw pitch
pitch = Pitch(pitch_color='black', line_color='white', stripe=False)
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

# Loop through filtered shots
for i in range(len(df_filtered)):
    shot_x = df_filtered.iloc[i]['x']  # Where the shot was taken from
    shot_y = df_filtered.iloc[i]['y']
    
    # If dataset has actual shot end locations, use them
    if 'end_x' in df_filtered.columns and 'end_y' in df_filtered.columns:
        goal_x = df_filtered.iloc[i]['end_x']
        goal_y = df_filtered.iloc[i]['end_y']
    else:
        goal_x = 120  # Assume goal center if end_x not available
        goal_y = 40

    # Set colors for each type of shot
    color_dict = {
        'MissedShots': 'red',
        'SavedShots': 'yellow',
        'BlockedShots': 'purple',
        'Goal': 'green'
    }
    
    shot_result = df_filtered.iloc[i]['result']
    shot_color = color_dict.get(shot_result, 'white')  # Default to white if unknown

    # Draw trajectory from shot location to actual shot end location
    plt.plot([shot_x, goal_x], [shot_y, goal_y], color=shot_color, linestyle='dashed', linewidth=1)

    # Mark the shot location
    plt.scatter(shot_x, shot_y, color=shot_color, s=50)

# Title
plt.title('Haaland Shot Map vs West Ham', color='White', size=20)


```
